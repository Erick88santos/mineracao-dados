Importando a biblioteca e o dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/rodrigoclira/mineracao-dados/main/datasets/house-price-train.csv')
```

```python
sns.histplot(train['SalePrice']);
```

Normalização é o processo de escalonamento os valores entre [0, 1]

Padronização é o processo de escalonamento das variáveis para ter as propriedades de uma distribuição gaussiana com média 0 e desvio padrão 1. 


Diferenças: 

* Quando padronização é usado, seus novos dados não são limitados a uma faixa de valor. 
* Use normalização quando você não sabe como é a distribuição dos seus dados ou você sabe que não se comporta como uma gaussiana. Use padronização se seus dados são gerados numa distribuição gaussiana. 
* As vezes quando a normalização não funciona, padronização funciona. 


## Método de Normalização 1 - Escalonando de forma simples

```python
train1 = train.copy()

print(train['SalePrice'].head())

train1['SalePrice'] = train1['SalePrice']/train1['SalePrice'].max()

print("\nnormalizado...")
print(train1['SalePrice'].head())
```

## Método de Normalização 1 - Min-Max

image.png

```python
train2 = train.copy()
print(train2['SalePrice'].head())

train2['SalePrice'] = (train2['SalePrice']-train2['SalePrice'].min())/(train2['SalePrice'].max()-train2['SalePrice'].min())

print("\nnormalizado...")
print(train2['SalePrice'].head())
```

## Padronização - Z-Score

image.png

```python
train3 = train.copy()
print(train3['SalePrice'].head())

train3['SalePrice'] = (train3['SalePrice']-train3['SalePrice'].mean())/train3['SalePrice'].std()
print("\nnormalizado...")
print(train3['SalePrice'].head())
```
